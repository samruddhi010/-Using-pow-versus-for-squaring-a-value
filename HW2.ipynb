```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data set
df = pd.read_csv('diabetic_data.csv')
```

```python
#shape of the data set
df.shape
```

```python
#data types for all the columns
df.dtypes
```

```python
#first 10 rows of the data
df.head(10)
```

```python
#last 15 rows of the data
df.tail(15)
```

```python
#summary of stats for the data set
df.describe()
```

```python
#plot gender attributes vs count as a bar chart
df['gender'].value_counts().plot(kind = 'bar')
plt.xlabel('Gender', labelpad=14)
plt.ylabel('Count of people', labelpad=14)
```

```python
#plot gender attributes vs count but now with a Race as a hue
fig = plt.figure(figsize=(10,8))
sns.countplot(x="gender", hue="race", data=df)
plt.show()
```

```python
#print the co-relation between all the attributes in a data set
corr = df.corr()
corr
```

```python
#using co-relation plot a heat map
plt.figure(figsize=(16,9))
sns.heatmap(corr, annot = True)
```
